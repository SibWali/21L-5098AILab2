```python
#Question2

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Read genome.txt
with open('genome.txt', 'r') as file:
   genome_sequence = file.read().strip()

genome_list = list(genome_sequence)
genome_length = len(genome_list)

#Generate helix coordinates
t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)
coordinates = np.column_stack((x, y, z))

#Assign colors (A=red, T=green, C=blue, G=yellow)
color_map = {'A': 'red', 'T': 'green', 'C': 'blue', 'G': 'yellow'}
colors = [color_map.get(nucleotide, 'gray') for nucleotide in genome_list]

#Create 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=colors, s=10)
ax.set_title('3D Helix Structure of Genome Sequence')
plt.show()
```

```python
#Question 3

import matplotlib.pyplot as plt
import numpy as np

#Load an image
img = plt.imread('/content/cardmethod.png')
img_array = np.array(img)

#Plot original image
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.imshow(img_array)
plt.title('Original Image')

#Rotate image
rotated_img = np.rot90(img_array)
plt.subplot(2, 2, 2)
plt.imshow(rotated_img)
plt.title('Rotated 90°')

#Flip image
flipped_img = np.fliplr(img_array)
plt.subplot(2, 2, 3)
plt.imshow(flipped_img)
plt.title('Flipped Horizontally')

#Grayscale conversion
gray_img = np.dot(img_array[..., :3], [0.299, 0.587, 0.114])
plt.subplot(2, 2, 4)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')

plt.tight_layout()
plt.show()
```

```python
#Question 4
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

#Load dataset
iris = load_iris()
X = np.array(iris.data)  # Features
Y = np.array(iris.target)  # Target

#Task 1: Calculate statistics
print("Mean:", np.mean(X, axis=0))
print("Median:", np.median(X, axis=0))
print("Std Dev:", np.std(X, axis=0))
print("Min:", np.min(X, axis=0))
print("Max:", np.max(X, axis=0))

#Extract sepal length and width
sepal_features = X[:, [0, 1]]

#Task 2: Visualizations
plt.figure(figsize=(15, 5))

#Scatter plot
plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Width')

#Histogram
plt.subplot(1, 3, 2)
plt.hist(X[:, 0], bins=20)
plt.xlabel('Sepal Length')
plt.title('Sepal Length Distribution')

#Line plot (sorted by petal length)
sorted_indices = np.argsort(X[:, 2])
plt.subplot(1, 3, 3)
plt.plot(X[sorted_indices, 2], X[sorted_indices, 3])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Width')

plt.tight_layout()
plt.show()
```
